```python
from sklearn.impute import KNNImputer
```

```python
numeric_cols = df.select_dtypes(include=['float', 'int']).columns
imputer = KNNImputer(n_neighbors=4)
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
df
```

```python
df['Processing.Method'] = df['Processing.Method'].str.lower().replace({
       'washed': 'washed',
       'wet': 'washed'
   })
df
```

```python
heat_map = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance']
sns.heatmap(df[heat_map].corr(), annot=True, cmap='coolwarm')

```

```python
sns.scatterplot(x='Altitude', y='Acidity', hue='Species', data=df)
plt.title("Scatter plot")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = df[['altitude_mean_meters']]
y = df['Acidity']
model = LinearRegression().fit(X, y)
print(f"Regression Equation: y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")
```

```python
sns.scatterplot(x='altitude_mean_meters', y='Acidity', data=df)
plt.plot(df['altitude_mean_meters'], model.predict(X), color='red')

```
